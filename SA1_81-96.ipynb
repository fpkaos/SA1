## 81.	Оценка соответствия ИБ организации установленным требованиям

**Оценка соответствия ИБ организации установленным требованиям** - любая деятельность, связанная с прямым или косвенным определением того, что выполняются или не выполняются соответствующие требования информационной безопасности в организации.

С помощью оценки соответствия ИБ измеряется правильность реализации процессов и защитных мер СОИБ организации и идентифицируются недостатки такой реализации.

В качестве эталона могут быть:
- требования законодательства РФ в области ИБ;
- отраслевые требования по ИБ;
- требования национальных и международных стандартов по ИБ.


## 82.	Этапы оценки соответствия  ИБ

Этапы оценки соответствия ИБ (оценка ИБ по эталону):
- выбор эталона;
- формирование критериев оценки ИБ на основе эталона;
- сбор свидетельств оценки ИБ;
- измерение параметров объекта оценки ИБ;
- формирование оценки ИБ.

## 83.	Риск-ориентированная оценка ИБ

**Риск-ориентированная оценка ИБ организации** представляет собой способ оценки, при котором рассматриваются риски ИБ, возникающие в информационной сфере организации, и сопоставляются существующие риски ИБ и принимаемые меры по их обработке. В результате должна быть сформирована оценка способности организации эффективно управлять рисками ИБ для достижения своих целей.

Основные этапы риск-ориентированной оценки информационной безопасности включают: идентификацию рисков ИБ, определение адекватных процессов менеджмента рисков и ключевых индикаторов рисков ИБ, формирование на их основе критериев оценки ИБ, сбор свидетельств оценки и измерение риск-факторов, формирование оценки ИБ.

Оцениваются:
- риски ИБ, присущие информационной сфере организации, (неотъемлемые риски ИБ);
- риски ИБ, связанные с внутренним контролем.

## 84.	Присущий (неотъемлемый) риск

**Присущий (неотъемлемый) риск** - опасность совершения существенно важной ошибки при ведении бизнес-операций в отсутствие внутреннего контроля.

Риски, связанные с недостаточной эффективностью механизмов внутреннего контроля, представляют собой опасность того, что меры внутреннего контроля окажутся не способны своевременно предотвратить или выявить искажения, которые могут повлечь за собой существенно важные последствия.

Риск-ориентированный подход даёт возможность экономить ресурсы, т.к. исходя из оценки неотъемлемого риска и риска контроля на этапе планирования можно выделить наиболее значимые (существенные) риски и при проведении проверки оценивать именно эти риски.

## 85.	Оценка ИБ на основе экономических показателей

Для проведения оценки ИБ в качестве критериев эффективности СОИБ используются, например, показатели совокупной стоимости владения [(ТСО, Total Cost of Ownership)](https://ru.wikipedia.org/wiki/%D0%A1%D0%BE%D0%B2%D0%BE%D0%BA%D1%83%D0%BF%D0%BD%D0%B0%D1%8F_%D1%81%D1%82%D0%BE%D0%B8%D0%BC%D0%BE%D1%81%D1%82%D1%8C_%D0%B2%D0%BB%D0%B0%D0%B4%D0%B5%D0%BD%D0%B8%D1%8F).

Под показателем TCO понимается сумма прямых и косвенных затрат на внедрение, эксплуатацию и сопровождение СОИБ.

Оценка на основе показателя TCO позволяет оценить затраты на информационную безопасность и сравнить ИБ организации с типовым профилем защиты, а также управлять затратами для достижения требуемого уровня защищенности.

Для управления затратами на обеспечение ИБ необходимо сравнить результаты ТСО для конкретной организации с информационной базой, созданной по результатам анализа показателей в других организациях аналогичного бизнеса.

Способ оценки ИБ требует создания общей информационной базы данных об эффективности СОИБ организаций схожего бизнеса и постоянной поддержки базы данных в актуальном состоянии. Такое информационное взаимодействие организаций, как правило, не соответствует целям бизнеса. Поэтому оценка ИБ на основе показателя TCO редко применяется.

## 86.	Цели оценки соответствия ИБ организации

- поиск несоответствий установленным требованиям ИБ;
- поддержка информированности руководства организации об ИБ;
- сравнение организаций с аналогичным бизнесом и сопоставимым размером;
- выявление влияния критических элементов (факторов) и их сочетания на ИБ организации.

## 87.	Значение аудита информационной безопасности

Одним из условий реализации целей деятельности любой организаций является обеспечение необходимого и достаточного уровня информационной безопасности (ИБ) их активов (в т. ч. информационных).

Основным из видов проверки уровня ИБ в организациях является аудит ИБ. Мировой опыт в области обеспечения ИБ определяет аудит ИБ как важнейший процесс в непрерывном цикле процессов менеджмента ИБ организации.

Основная цель аудита ИБ есть «оценка соответствия ИБ организаций требованиям выбранных стандартов по ИБ и требованиям  регуляторов в области ИБ (compliance)».

СОИБ должна быть **определена, спланирована и регламентирована** в организации. Однако даже правильно выстроенные процессы и используемые защитные меры в силу объективных причин со временем имеют тенденцию к ослаблению своей эффективности. Это неминуемо ведет к деградации системы защиты и возрастанию рисков нарушения ИБ

Для поддержания системы защиты на должном уровне в качестве оперативной меры используется мониторинг событий и инцидентов в СИБ. Менеджмент событий и инцидентов безопасности, полученных в результате мониторинга, позволяет избежать деградации и обеспечить требуемый уровень безопасности активов.

Для оценки состояния ИБ защищаемого актива и выявления признаков деградации используемых защитных мер проводится оценка (самооценка) соответствия системы требованиям выбранных стандартов»

## 88.	Принципы аудита ИБ

### 1. Независимость аудита 

Аудиторы **независимы** в своей деятельности и **неответственны** за деятельность, которая подвергается аудиту ИБ. Независимость является основанием для **беспристрастности** при проведении аудита ИБ и **объективности** при формировании заключения по результатам аудита ИБ

Принцип частично применим для самооценки

Самооценка может проводиться лицами, **ответственными** за проверяемую деятельность, что потенциально несет **риски конфликта интересов**

### 2. Полнота аудита ИБ

Аудит ИБ должен охватывать все области аудита ИБ, соответствующие аудиторскому заданию. Кроме того, полнота аудита ИБ определяется достаточностью затребованных и предоставленных материалов, документов и уровнем их соответствия поставленным задачам. Полнота аудита ИБ является необходимым условием для формирования объективных заключений по результатам аудита ИБ.

Принцип частично применим для самооценки

Самооценка должна охватывать все области ИБ, соответствующие плану самооценки, но в то же время план самооценки может ограничиваться теми элементами деятельности, продуктов и услуг организации, которые руководство считает наиболее приоритетными. Кроме того, полнота самооценки ИБ может быть большей по отношению к аудиту ИБ в силу большей возможности по использованию необходимых, в т.ч чувствительных, материалов (свидетельств самооценки).


### 3. Оценка на основе свидетельств аудита

При периодическом проведении аудита ИБ оценка на основе свидетельств аудита ИБ является единственным способом, позволяющим получить повторяемое заключение по результатам аудита ИБ, что повышает доверие к такому заключению. Для повторяемости заключения свидетельства аудита ИБ должны быть проверяемыми

По сравнению с аудитом ИБ самооценка ИБ может значительно превосходить по количеству и качеству свидетельств.


### 4. Достоверность аудита ИБ

У аудиторов должна быть уверенность в достоверности свидетельств аудита ИБ. Доверие к документальным свидетельствам аудита ИБ повышается при подтверждении их достоверности третьей стороной или руководством организации. Доверие к фактам, полученным при опросе сотрудников проверяемой организации, повышается при подтверждении данных фактов из различных источников. Доверие к фактам, полученным при наблюдении за деятельностью проверяемой организации в области ИБ, повышается, если они получены непосредственно при функционировании проверяемых процедур или процессов.

Принцип частично применим для самооценки

Заинтересованность лиц в оценке своей деятельности предполагает использование только достоверных свидетельств. В то же время, такие методы получения свидетельств как опрос и наблюдение за деятельностью мало применимы при проведении самооценки ИБ.

### 5. Компетентность

Доверие к процессу и результатам аудита ИБ зависит от компетентности тех, кто проводит аудит ИБ, и от этичности их поведения. Компетентность базируется на способности аудитора применять знания и навыки

Если при проведении аудита компетентность достигается на этапе выбора аудиторской организации (посредством конъюнктурного анализа)

### 6. Этичность поведения

Этичность поведения подразумевает ответственность, не подкупность, умение хранить тайну, беспристрастность.

Трудно применить беспристрастность при оценке своей деятельности

## 89.	Требования к этапам проведения аудита

Работы по проведению аудита ИБ организаций должны включать следующие этапы:
- подготовка к проведению аудита ИБ;
- анализ документов;
- проведение аудита ИБ на месте;
- подготовка, утверждение и рассылка отчета по аудиту ИБ;
- завершение аудита ИБ

## 90.	Подготовка к проведению аудита

Подготовка к проведению аудита ИБ должна начинаться с согласования и заключения договора на проведение аудита ИБ. В договор рекомендуется включать:
- область аудита ИБ;
- ответственность руководства проверяемой организации за подготовку и предоставление необходимых свидетельств аудита ИБ;
- требования к отчету аудита ИБ;
- порядок взаимодействия представителей проверяющей и проверяемой организаций;
- порядок сбора необходимых свидетельств аудита;
- цена проведения аудита ИБ;
- порядок привлечения к работе по каким-либо вопросам аудита ИБ других проверяющих организаций и(или) технических экспертов;
- необходимые ограничения ответственности проверяющей организации.

## 91.	План аудита

План аудита ИБ должен, как минимум, включать следующую информацию:
- цели аудита ИБ;
- критерии аудита ИБ;
- объекты, подвергающиеся оценке;
- показатели, которые должны быть оценены;
- порядок сбора свидетельств аудита;
- порядок анализа данных и форма отчетности;
- роли и обязанности соответствующих причастных сторон;
- пересмотр защитных мер по результатам аудита ИБ.

## 92.	Программа аудита

План деятельности по проведению одного или нескольких аудитов информационной безопасности (и других проверок информационной безопасности), запланированных на конкретный период времени и направленных на достижение конкретной цели.
Примечание. Программа аудита информационной безопасности включает всю деятельность, не обходимую для планирования, проведения, контроля, анализа и совершенствования аудитов информационной безопасности (и других проверок информационной безопасности)

Программа аудита ИБ включает деятельность, необходимую для планирования и организации определенного количества аудитов, их контроля, анализа и совершенствования, а также обеспечения их ресурсами, необходимыми для эффективного и результативного проведения аудитов ИБ в заданные сроки.
Программа аудита ИБ разрабатывается организацией. В общем случае могут быть разработаны несколько программ аудита ИБ

>Программа аудита определяет процессы менеджмента аудита для определенного количества аудитов

Программа аудита ИБ может включать, например, следующие виды проверок ИБ организации:
- аудит ИБ организации;
- самооценка ИБ, проводимая ежегодно;
- аудиты по сертификации и инспекционные аудиты.

Для успешного внедрения программы аудита ИБ необходимо:
- Выделить необходимые ресурсы;
- Распределить роли и полномочия в рамках процесса;
- Определить и задокументировать порядок формирования, сбора и хранения свидетельств аудита (самооценки) ИБ, периодичность их проведения и порядок хранения его результатов;
- Организовать обучение для сотрудников организации, участвующих в процессе аудита (самооценки).

### PDCA

1. Разработка программы аудита должна включать в себя:
    - определение целей программы;
    - определение объема программы;
    - определение необходимых финансовых, информационных и людских ресурсов для ее реализации;
    - согласование программы с ее участниками и утверждение на уровне руководства организации.

2. Внедрение программы аудита должно включать в себя:
    - доведение программы аудита ИБ до участвующих в ее реализации сторон;
    - планирование аудита (самооценки) ИБ;
    - проведение аудитов (самооценок) ИБ в соответствии с программой аудита ИБ;
    - анализ и утверждение отчетов по результатам аудита (самооценки) ИБ;
    - определение действий по результатам аудита (самооценки) ИБ.

3. В организации должны проводиться контроль внедрения программы аудита ИБ, анализ достижения целей программы аудита ИБ и определение возможностей для ее совершенствования. Контроль и анализ программы аудита ИБ должны включать в себя:
    - проверку возможностей рабочих групп, служб или лиц по реализации аудита (самооценки) ИБ;
    - анализ достижения целей аудита (самооценки) ИБ и программы аудита ИБ в целом;
    - анализ отчетов и заключений по результатам аудита (самооценки) ИБ.

4. Совершенствование программы аудита ИБ состоит в определении корректирующих и превентивных действий по совершенствованию программы аудита ИБ, включающих в себя:
    - пересмотр и корректировку сроков проведения аудита (самооценок) ИБ и необходимых ресурсов (повышение эффективности процесса);
    - улучшение методов подготовки свидетельств аудита (самооценки) ИБ (повышение результативности процесса)


## 93.	Анализ документации

До проведения аудита ИБ на месте рабочей группой должен быть проведен анализ требуемой документации с целью:
- определения соответствия положений, отраженных в документации, критериям аудита ИБ;
- более глубокого знакомства с организацией и областью оценки;
- оценки соответствия требованиям законодательства;
- экономии ресурсов проверяемой организации;
- выявления возможностей для приведения в соответствие и дальнейшего улучшения.

## 94.	Проведение аудита на месте

Проведение аудита ИБ на месте должно включать следующие работы:
- планирование аудита на месте
- проведение вступительного совещания;
- сбор свидетельств аудита ИБ и другой необходимой информации;
- оценка свидетельств аудита ИБ;
- подготовка заключения по результатам аудита ИБ;
- проведение заключительного совещания.

В зависимости от границ работ могут быть запланированы несколько этапов работ на местах

Основными источниками свидетельств аудита ИБ должны являться:
- документы проверяемой организации и третьих лиц, относящиеся к обеспечению ИБ организации;
- устные высказывания и письменные ответы сотрудников проверяемой организации в процессе проводимых опросов;
- результаты наблюдений аудиторов за деятельностью организации в области ИБ.
- результаты использования специальных технических средств. 

Основными методами получения свидетельств аудита ИБ должны являться:
- проверка и анализ документов, касающихся обеспечения ИБ организации;
- визуальный осмотр и наблюдение за деятельностью организации в области ИБ;
- опрос сотрудников проверяемой организации и независимой (третьей) стороны.
- результаты использования специальных технических средств

По степени достоверности (от наибольшей к наименьшей) свидетельства аудита ИБ делятся следующим образом:
- свидетельства, полученные от третьей стороны в письменном виде;
- свидетельства, полученные от проверяемой организации и подтвержденные третьей стороной в письменном виде;
- свидетельства, полученные в ходе проведения аудиторских процедур (наблюдения за деятельностью, анализа данных системы мониторинга ИБ и т.д.);
- свидетельства, полученные в форме документов;
